```python

import pandas as pd
load = pd.read_csv("best_buy_laptops_2024.csv")
load
```

```python
load.info()
```

```python
limpar = load.drop(['depth','width','features/0/description',
                    'features/1/description',
                    'model',
                    'aggregateRating/reviewCount',
                    'aggregateRating/ratingValue'], axis = 1)
limpar
```

```python
limpeza_full = limpar.dropna()
limpeza_full
```

```python
analisando1 = limpeza_full["offers/price"].describe()
analisando1
```

```python
analisando2 = limpeza_full["brand"].describe()
analisando2
```

```python
analisando1.plot.bar()
```

```python
analisando1.plot.hist()
```

```python
import matplotlib.pyplot as plt
```

```python
limpeza_full['brand'].head(10).value_counts().sort_values(ascending=False).plot.bar()

# Adicione um título ao gráfico
plt.title('Produtos Mais Vendidos')

# Adicione rótulos aos eixos
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')

# Mostre o gráfico
plt.show()
```
